# Two age groups with mixing

This scenario creates two groups of people. All the people in one group is 30 years old, the others are 80. Restriction values can be set on the individual groups but also between them. Here we will demo restricting movment between the two.

```python
from epidexus import EpidexusModel
from epidexus.scenarios.twoages import TwoAges

from datetime import datetime, timedelta
import numpy.random

young_poulation_size = 100
old_poulation_size = 100
infected_young = 10
infection_rate = 0.5
start_date = datetime(2020, 3, 31)
scenario = TwoAges(young_poulation_size, old_poulation_size, infection_rate, infection_rate, start_date, num_infected_young=infected_young)
scenario.set_seed(0)
```

## Simulate

Ten young has been infected, no old are infected. We will run the simulation for 60 days, and we see that because the restriction is one on the young to old visits, the old people does not get infected.

```python
scenario.restriction_young_to_old = 1.0
scenario.simulate(timedelta(days=60))
from epidexus.plotting import seir_stackplot
seir_stackplot(scenario.sim_model)
```
